```python
import pandas as pd

def parse_date(x):
    try:
        return pd.to_datetime(x, format='%m/%d/%Y')
    except:
        try:
            return pd.to_datetime('1899-12-30') + pd.to_timedelta(int(x), 'D')
        except:
            return pd.NaT
```

```python
df['OrderDate'] = df['OrderDate'].apply(parse_date)
df['ShipDate']  = df['ShipDate'].apply(parse_date)
```

```python
orders = pd.read_csv("/content/orders.csv")
customers = pd.read_csv("/content/customers.csv")
products = pd.read_csv("/content/products.csv")
```

```python
orders['OrderDate'] = orders['OrderDate'].apply(parse_date)
orders['ShipDate'] = orders['ShipDate'].apply(parse_date)
```

```python
order_summary = (
    orders
    .merge(customers, on="CustomerID", how="left")
    .merge(products, on="ProductID", how="left")
    .groupby(
        ['OrderID', 'OrderDate', 'ShipDate', 'CustomerName', 'Region', 'Segment', 'Category', 'SubCategory'],
        as_index=False
    )
    .agg(
        TotalSales = ('Sales', 'sum'),
        TotalProfit = ('Profit', 'sum'),
        TotalQuantity = ('Quantity', 'sum'),
        ProfitMargin = ('Profit', lambda x: round(x.sum() / max(orders.loc[x.index,'Sales'].sum(), 1), 3)),
        AvgDiscount = ('Discount', 'mean')
    )
)
```

```python
order_summary['AvgDiscount'] = order_summary['AvgDiscount'].round(3)
```

```python
output_path = "/content/order_summary_cleaned.csv"
order_summary.to_csv(output_path, index=False)
```
